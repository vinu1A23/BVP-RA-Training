```python
import math
import numpy as np
from matplotlib import pyplot as plt

#sigmoid
def sigmoid(x):
  return 1.0/(1.0+math.exp(-x))
```

```python
#tanh

def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
def relu(x):
  return max(0.0,x)
```

```python
#Leaky Relu
def leaky_relu(x):
  return max(0.0*x,x)
```

```python
input = list(np.linspace(-10,10,100))
```

```python
output=[sigmoid(x) for x in input]
plt.plot(input, output, c="r")
```

```python
output2=[tanh(x) for x in input]
plt.plot(input, output2, c="y")
```

```python
output3=[relu(x) for x in input]
plt.plot(input, output3, c="g")
```

```python
output4=[leaky_relu(x) for x in input]
plt.plot(input, output4, c="m")
```
