```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Sigmoid Activation Function



```python
# Range: 0, 1
x = np.linspace(-10,10,50)
```

```python
def sigmoid_fun(x):
  return 1/(1 + np.exp(-x))
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,sigmoid_fun(x))
plt.ylabel('Output values')
plt.xlabel('Inputs')
plt.title('Sigmoid Function')
```

Tanh Activation Function

```python
# Range: -1, 1
x = np.linspace(-20,30,100)
```

```python
def tanh_fun(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,tanh_fun(x),c='orange')
plt.ylabel('Output values')
plt.xlabel('Inputs')
plt.title('Tanh Function')
```

RECTIFIED LINEAR UNIT (RELU) ACTIVATION FUNCTION 

```python
# Range: 0, +infinity
x = np.linspace(-20,30,100)
```

```python
def relu_fun(x):
  value = [max(0, inp) for inp in x]
  return np.array(value)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,relu_fun(x),c='yellow')
plt.ylabel('Output values')
plt.xlabel('Inputs')
plt.title('Relu Function')
```

LEAKY RELU ACTIVATION FUNCTION

```python
# Range: -infinity, +infinity
x = np.linspace(-100,30,100)
```

```python
def lrelu_fun(x):
  value = [max(0.01*inp, inp) for inp in x]
  return np.array(value)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,lrelu_fun(x),c='black')
plt.ylabel('Output values')
plt.xlabel('Inputs')
plt.title('Leaky Relu Function')
```
