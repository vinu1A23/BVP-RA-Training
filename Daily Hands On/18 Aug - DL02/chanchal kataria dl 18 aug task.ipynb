```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import math
```

```python
#sigmoid (0,1)

def sigmoid(x):
  return 1.0/(1.0+math.exp(-x))
```

```python
#tanh (-1,1)

def tanh(x):
  s=math.exp(x)
  t=math.exp(-x)
  return (s-t)/(s+t)
```

```python
tanh(56.5)
```

```python
tanh(0)
```

```python
tanh(-56.5)
```

```python
tanh(1)
```

```python
tanh(-76)
```

```python
#leaky relu
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
#graph of sigmoid


def sigmoid(x):
  sig = 1/(1+np.exp(-x))
  return sig
x = np.linspace(-10,10,100)
plt.plot(x,sigmoid(x),c='r')
```

```python
#graph of tanh


def tanh(x):
  th = ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
  return th
x = np.linspace(-10,10,100)
plt.plot(x,tanh(x),c='r')
```

```python
#graph of relu


def relu(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
x_data = np.linspace(-10,10,100)
y_data = relu(x_data)
plt.plot(x_data, y_data)
```
