```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input = np.linspace(-10,10,100)
```

# 1.relu


```python
def relu(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
```

```python
func = np.vectorize(relu)
plt.plot(input, relu(input))
```

## 2.sigmoid

```python
import math

def sigmoid(x):
  return 1/(1+ np.exp(-x))
```

```python
plt.plot(input, sigmoid(input), c='r')
```

# 3.tanh


```python
#tanh (-1,1)

def tanh(x):
  s=np.exp(x)
  t=np.exp(-x)
  return (s-t)/(s+t)

```

```python
plt.plot(input, tanh(input))
```

## 4.leaky relu

```python
def leaky_ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)
```

```python
func = np.vectorize(leaky_ReLU)
plt.plot(input, leaky_ReLU(input))
```
