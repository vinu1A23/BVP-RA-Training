```python
#sigmoid(0,1)
import math

def sigmoid(x):
  return 1.0/(1.0+math.exp(-x))
```

```python
sigmoid(1)
```

```python
#tanh (-1,1)

def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x)) 
```

```python
tanh(5)
```

```python
def relu(x):
  return max(0.0,x)
```

```python
relu(18)
```

```python
#leaky relu
def leaky_relu(x):
   return max(0.1*x,x)
```

```python
leaky_relu(-0.45)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
```

```python
#sigmoid graph
def sigmoid(x):
  sig = 1/(1+np.exp(-x))
  return sig
```

```python
plt.plot(x,sigmoid(x),c='r')
```

```python
#tanh graph
x = np.linspace(-10,10,30)
```

```python
def tanh(x):
  th = ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
  return th
```

```python
plt.plot(x,tanh(x),c='r')
```

```python
#reLU graph
def relu(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
```

```python
x_data = np.linspace(-10,10,100)
y_data = relu(x_data)
```

```python
plt.plot(x_data, y_data)
```
