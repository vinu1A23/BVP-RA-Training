```python
import math
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sigmoid {range: [0, 1], threshold: 0.5}
def sigmoid(x):
  return 1/(1+math.exp(-x))
```

```python
sigmoid(-56)
```

```python
#tanh {range: [-1, 1], threshold: 0}

def tanh(x):
  return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))
```

```python
tanh(0.1)
```

```python
#ReLU {range: [0, infinity), threshold: 0}

def ReLU(x):
  return max(0.0, x)
```

```python
#Leaky ReLU

def leaky_ReLU(x):
  return max(0.1*x, x)
```

```python
leaky_ReLU(-0.45)
```

```python
X = np.arange(-10, 10, 0.01)
print(X)
```

```python
# sigmoid graph
sigmoid_arr = []
for i in X:
  sigmoid_arr.append(sigmoid(i))
# print(sigmoid_arr)
plt.plot(X, sigmoid_arr)
plt.show
```

```python
# tanh graph
tanh_arr = []
for i in X:
  tanh_arr.append(tanh(i))
plt.plot(X, tanh_arr)
plt.show
```

```python
# ReLU graph
ReLU_arr = []
for i in X:
  ReLU_arr.append(ReLU(i))
plt.plot(X, ReLU_arr)
plt.show
```

```python
# Leaky ReLU graph
leaky_arr = []
for i in X:
  leaky_arr.append(leaky_ReLU(i))  
plt.plot(X, leaky_arr)
plt.show
```
