```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('insurance_data.csv')
df.head()
```

```python
df.shape
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
feature_set=df.drop('bought_insurance',axis=1)
labels=np.array(df['bought_insurance'])
labels=labels.reshape(28,1)
```

```python
np.random.seed(50)
weights=np.random.rand(2,1)
bias=np.random.rand(1)
lr=0.001
```

```python
for epoch in range(30000):
    inputs=feature_set
    
    y=np.dot(feature_set,weights)+bias
    z=sigmoid(y)
    
    error=z-labels
    print(error.sum())
    
    dcost_dpred=error
    dpred_dz=sigmoid_der(z)
    
    z_delta=dcost_dpred*dpred_dz
    
    inputs=feature_set.T
    weights-=lr*np.dot(inputs,z_delta)
    
    for num in z_delta:
        bias-=lr*num
```

```python
#person with age 45 and affordibility 1
single_point=np.array([45,1])
result=sigmoid(np.dot(single_point,weights)+bias)
print(result)
```
