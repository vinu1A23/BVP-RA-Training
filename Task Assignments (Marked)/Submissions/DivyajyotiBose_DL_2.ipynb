```python
import math as m
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return(1/(1+np.exp(-x)))
```

```python

x=np.linspace(-5,5,100)
plt.plot(x,sigmoid(x),c="k")
plt.ylabel("Sigmoid")
plt.xlabel("Input")
```

```python
def tanh(x1):
    return(m.exp(x1)-m.exp(-x1))/(m.exp(x1)+m.exp(-x1))
```

```python
x1=np.linspace(-np.pi,np.pi,30)
y1=np.vectorize(tanh)
plt.plot(x1,y1(x1),c="b")
```

```python
def relu(x):
    return(max(0,x,key=int))
```

```python
x=np.arange(-20,20,5)
y=np.vectorize(relu)
plt.plot(x,y(x),c="k")
plt.ylabel("RELU")
plt.xlabel("Input")
```
