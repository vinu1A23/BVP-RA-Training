```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
import seaborn as sn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/harshit2000/BVP-RA-Training/main/Assignment%201/Task/insurance_data.csv")
print(df)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
data = df.drop(['bought_insurance'], axis=1)
```

```python
feature_set = data.to_numpy()
labels = df.bought_insurance.to_numpy()
labels = labels.reshape(28,1)
print(feature_set)
print(labels)
```

```python
np.random.seed(1)
weights = np.random.rand(2,1)
bias = np.random.rand(1)
lr=0.05
```

```python
for epoch in range(20000):
  inputs = feature_set

  XW = np.dot(feature_set, weights) + bias

  z=sigmoid(XW)

  error = z-labels

  print(error.sum())

  dcost_dpred = error
  dpred_dz = sigmoid_der(z)

  z_delta = dcost_dpred * dpred_dz
  inputs = feature_set.T
  weights -= lr * np.dot(inputs, z_delta)

  for num in z_delta:
    bias -=lr*num
```

```python
single_point = np.array([28,1])
result = sigmoid(np.dot(single_point, weights)+bias)
print (result)
```
