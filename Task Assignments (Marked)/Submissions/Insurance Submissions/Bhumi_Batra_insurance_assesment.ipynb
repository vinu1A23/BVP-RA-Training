```python
from sklearn.model_selection import train_test_split
```

```python
X= data.drop("bought_insurance",axis=1)
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(X, Y , test_size=0.1)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU, PReLU, ELU
```

```python
classifier= Sequential()
```

```python
classifier.add(Dense(units=10, kernel_initializer= 'he_normal',activation='relu',input_dim=2))
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
 model= classifier.fit(xtrain,ytrain, validation_split=0.33, batch_size= 10, epochs=20000)
```

```python
ypred = classifier.predict(xtest)
```

```python
ypred= (ypred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm= confusion_matrix(ytest,ypred)
```
