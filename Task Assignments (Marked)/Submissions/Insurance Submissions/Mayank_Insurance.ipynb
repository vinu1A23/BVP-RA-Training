```python
import pandas as pd
import numpy as np
```

```python
url="https://raw.githubusercontent.com/harshit2000/BVP-RA-Training/main/Assignment%201/Task/insurance_data.csv"
d = pd.read_csv(url)
d.head()
```

```python
feature_set = d.drop('bought_insurance',axis=1)
```

```python
labels = d['bought_insurance']
```

```python
labels = labels.array.reshape(28,1)
```

```python
np.random.seed(45)
```

```python
weights = np.random.rand(2,1)
```

```python
bias = np.random.rand(1)
```

```python
lr = 0.001
```

```python
def sigmoid (x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der (x):
  return sigmoid(x)*(1-sigmoid(x))# 1/(1-e^-x)^2
```

```python
for epoch in range(20000):
  input = feature_set
  #feed forward step 1
  xw = np.dot(feature_set,weights) + bias
  #feed forward step 2
  z = sigmoid(xw)

  #back propogation
  error = z - labels
  print(error.sum())

  #back propogation step 2
  dcost_dpred = error # d cost at predicted value

  dpred_dz = sigmoid_der(z) #predicted dz

  z_delta = dcost_dpred * dpred_dz #delta of dcost and dpred

  inputs = feature_set.T # feature set given to inputs
  weights -= lr * np.dot(inputs,z_delta)

  for num in z_delta:
    bias -= lr * num
```
