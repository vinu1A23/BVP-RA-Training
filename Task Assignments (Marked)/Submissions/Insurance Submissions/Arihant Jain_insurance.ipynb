```python
features = np.array(df[['age', 'affordibility']])
print(features)
print(features.shape)
```

```python
labels = np.array(df['bought_insurance']).reshape(28, 1)
print(labels)
```

```python
np.random.seed(10)
```

```python
weights = np.random.randn(2, 1)
bias = 0.1*np.random.randn(1)
lr = 0.0005
```

```python
def sigmoid(x):
  return 1.0/(1.0 + np.exp(-x))

def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(70000):
  inputs = features

  # feedforward
  XW = np.dot(inputs, weights) + bias
  z = sigmoid(XW)

  #  back prop 1
  error = z - labels
  print(error.sum())

  dloss_dpred = error
  dpred_dz = sigmoid_der(z)
  zdelta = dloss_dpred*dpred_dz

  weights -= lr*np.dot(inputs.T, zdelta)
  for num in zdelta:
    bias -= lr*num
    # print(bias)
```
