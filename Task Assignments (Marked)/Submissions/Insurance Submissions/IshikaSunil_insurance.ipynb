```python
features=np.array(df[['age','affordibility']])
print(features)
```

```python
labels=np.array(df['bought_insurance']).reshape(28,1)
print(labels)
```

```python
np.random.seed(1)
weights=np.random.rand(2,1)
bias=np.random.rand(1)
lr=0.0005
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(30000):

  # feedforward
  xw= np.dot(features, weights)+bias
  z=sigmoid(xw)

  # backprop
  error=z-labels
  print(error.sum())

  dloss_dpred=error
  dpred_dz=sigmoid_der(z)
  z_delta= dloss_dpred*dpred_dz

  inputs= features.T
  weights-=lr*np.dot(inputs,z_delta)
  for num in z_delta:
    bias-=lr*num
```

```python
single_point= np.array([54,1])
result=sigmoid(np.dot(single_point, weights)+ bias)
print(result)
```
