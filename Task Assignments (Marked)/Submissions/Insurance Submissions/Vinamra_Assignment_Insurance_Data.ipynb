```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/VinamraAg/BVP-RA-Training/main/Assignment%201/Task/insurance_data.csv")
df
```

```python
a = df.drop(columns = 'bought_insurance')
b = df['bought_insurance']
```

```python
feature_set = a.to_numpy()
labels = b.to_numpy()
labels = labels.reshape(28,1)
```

```python
np.random.seed(35)
weights = np.random.rand(2,1)
bias = np.random.rand(1)
lr = 0.001
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_der(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
for epoch in range(20000):
  inputs = feature_set

  XW = np.dot(feature_set, weights) + bias

  z = sigmoid(XW)

  error = z - labels

  print(error.sum())

  dcost_dpred = error
  dpred_dz = sigmoid_der(z)

  z_delta = dcost_dpred * dpred_dz

  inputs = feature_set.T
  weights -= lr * np.dot(inputs, z_delta)

  for num in z_delta:
        bias -= lr * num
```

```python
single_instance = np.array([34,0])
result = sigmoid(np.dot(single_instance, weights) + bias)
print(result)
```
