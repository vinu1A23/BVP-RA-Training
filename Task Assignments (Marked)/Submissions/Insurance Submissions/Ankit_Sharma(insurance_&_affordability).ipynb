```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/harshit2000/BVP-RA-Training/main/Assignment%201/Task/insurance_data.csv")
df
```

```python
input = df.drop(columns = 'bought_insurance')
target = df['bought_insurance']
feature_set = input.to_numpy()
labels = target.to_numpy()
labels = labels.reshape(28,1)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
def sigmoid_der(x):
  return sigmoid(x) * (1-sigmoid(x))
```

```python
np.random.seed(36)
weights = np.random.rand(2,1)
bias=np.random.rand(1)
lr = 0.0009
```

```python
for epoch in range(25000):
  inputs = feature_set

  XW = np.dot(feature_set,weights) + bias

  z = sigmoid(XW)

  error = z - labels

  print(error.sum())

  dcost_dpred = error

  dpred_dz = sigmoid_der(z)

  z_delta = dcost_dpred * dpred_dz

  inputs = feature_set.T

  weights -= lr * np.dot(inputs,z_delta)

  for num in z_delta:
    bias -= lr * num
```

```python
single_point = np.array([68,1])
result = sigmoid(np.dot(single_point,weights)+bias)
print(result)
```
