```python
import numpy as np
import pandas as pd
import math
```

```python
data = pd.read_csv("hi.csv")
print(data)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
x = data.drop(columns='bought_insurance')
y=data['bought_insurance']
x
```

```python
feature=x.to_numpy()
labels=y.to_numpy()
labels=labels.reshape(28,1)
np.random.seed(36)
weights=np.random.rand(2,1)
bias=np.random.rand(1)
lr=0.0005
```

```python
for epoch in range(30000):
    inputs=feature
    #feed forwarding step 1
    xw=np.dot(feature,weights)+bias
    #feed forwarding step 2
    z=sigmoid(xw)
    #back propagation step 1
    error=z-labels
    print(error.sum())
    #back propagation step 2
    dcost_dpred=error
    dpred_dr=sigmoid_der(z)

    x_delta=dcost_dpred*dpred_dr
    inputs=feature.T
    weights-=lr*np.dot(inputs,x_delta)
    
    for num in x_delta:
        bias-=lr*num
```

```python
#person who is 55 years old and 1 represents they can afford insurance
z = np.array([55,1])
result = sigmoid(np.dot(z,weights)+bias)
print(result)
```

```python
#person who is 45 years old and 0 represents they can't afford insurance
z = np.array([45,0])
result = sigmoid(np.dot(z,weights)+bias)
print(result)
```
