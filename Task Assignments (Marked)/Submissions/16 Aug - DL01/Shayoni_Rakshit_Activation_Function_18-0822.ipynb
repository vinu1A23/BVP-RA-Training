```python

import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
input = np.linspace(-10, 10)
```

```python
#Sigmoid [0,1]
def sigmoid(x):
   return 1.0/(1.0+np.exp(-x))
```

```python
sigmoid(0.23)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
plt.plot(input,sigmoid(input),c="r")
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
# Tanh Function: [-1,1]

def tanh(x):
  a = np.exp(x)
  b = np.exp(-x)
  return (a - b) / (a + b)
```

```python
tanh(1)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
plt.plot(input,tanh(input))
plt.title('Hyperbolic Tan Activation Function')
plt.show()
```

```python
#Relu Function
def relu(x):
  return np.maximum(0,x)
```

```python
relu(-100)
```

```python
relu(1000)
```

```python
x1=np.linspace(-5,5)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,4))
plt.plot(x1,relu(x1))
plt.title('Relu Function')
plt.show()
```

```python
# Leaky ReLu
def leaky_relu(x):
  return np.maximum(0.1*x,x)
```

```python
leaky_relu(-0.45)
```

```python
# Leaky ReLu
plt.style.use('seaborn')
plt.figure(figsize=(8,4))

arr = np.linspace(-5, 5)
plt.plot(arr, leaky_relu(arr))
plt.title('Leaky ReLU Activation Function')
plt.show()
```
