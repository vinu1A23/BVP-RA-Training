```python
import numpy as np
import matplotlib.pyplot as plt
import math
x=np.linspace(-10,10,200)
```

```python
# Sigmoid function
def sigmoid(x):
    return 1/(1 + np.exp(-x));
```

```python
sigmoid(1.3)
```

```python
plt.plot(x,sigmoid(x))
```

```python
#Tanh function
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
tanh(3)
```

```python
plt.plot(x,tanh(x))
```

```python
# Relu function
def relu(x):
    return np.maximum(0,x)
```

```python
relu(-2)
```

```python
relu(5)
```

```python
plt.plot(x,relu(x))
```
