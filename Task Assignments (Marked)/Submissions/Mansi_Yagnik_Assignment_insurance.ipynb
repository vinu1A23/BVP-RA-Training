```python
feature_set = data.iloc[:,[0,1]]
labels = data.iloc[:, -1]
```

```python
labels = labels.array.reshape(28,1)
```

```python
np.random.seed(45)
weight=np.random.rand(2,1)
bias=np.random.rand(1)
lr=0.01
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(20000):
  input = feature_set

  xw = np.dot(feature_set,weight) + bias

  z = sigmoid(xw)

  error =z-labels
  

  print(error.sum())

  
  dcost_dpred = error 

  dpred_dz = sigmoid_der(z) 

  z_delta = dcost_dpred * dpred_dz 

  inputs = feature_set.T 
  weight -= lr * np.dot(inputs,z_delta)

  for num in z_delta:
    bias -= lr * num
```
