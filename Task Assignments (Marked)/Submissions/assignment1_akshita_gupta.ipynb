```python
features=np.array(data[['age','affordibility']])
#print(features)
```

```python
labels=np.array(data.bought_insurance).reshape(28,1)
#labels
```

```python
np.random.seed(1)
weights=np.random.rand(2,1)
bias=np.random.rand(1)
lr=0.0005
#bias
#weights
#np.dot(features,weights)
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
   return sigmoid(x) * (1-sigmoid(x))
```

```python
for epoch in range(30000):
  
  #feed forward step 1
  XW=np.dot(features,weights)+bias

  #feed forward step 2 (activation function)
  z=sigmoid(XW)

  #back propogation step 1
  error=z-labels
  print(error.sum())

  #back propogation step 2
  dcost_dpred=error
  dpred_dz=sigmoid_der(z)

  z_delta=dcost_dpred*dpred_dz

  inputs=features.T
  weights -= lr * np.dot(inputs, z_delta)

  for num in z_delta:
    bias -= lr * num
```
