```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))

```

```python
age=df_insurance["age"]#.iloc[:,0]
affordability=df_insurance["affordibility"]#.iloc[:,1]
#df_new_independent=age+affordability
bought=df_insurance["bought_insurance"]#.iloc[:,2]
df_new_independent=np.array(df_insurance.iloc[:,:2])

#df_new_independent
bought_np=np.array(df_insurance["bought_insurance"])#.iloc[:,2])
#bought_np
bought_np=bought_np.reshape(28,1)
```

```python

np.random.seed(2)
weights = np.array([[0.5],[0.2]])
#weig
bias = np.random.rand(1)
lr = 0.006
#weights


```

```python
for epochs in range(20000):
  inputs=df_new_independent

  XW=np.dot(df_new_independent,weights)+ bias
  #print(XW)
  z=sigmoid(XW)
 
  #backward propagation step 1
  #error=z-np.array(bought)
  error=z-bought_np
  print(error.sum())
  
  #z_label
  #dprederror=error 
  #dcostprederror=dprederror*sigmoid_der(dprederror) #sigmoid_der(dcost_dpred) is dpred_dz
  
  #backward propagation step 2
  dcost_dpred=error
  dpred_dz=sigmoid_der(z)
  
  z_delta=dcost_dpred*dpred_dz
  #print(z_delta)
  inputs=df_new_independent.T 
  #inputs.reshape
  
  weights-=lr*np.dot(inputs,z_delta)#dcostprederror

  
  for i in z_delta:
    bias -= lr* i

  #inputs=df_new_independent.T
```
